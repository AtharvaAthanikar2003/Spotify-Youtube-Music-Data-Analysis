Q4. How do the Average Views, Likes, Comments are compared between different Album Types ?

```python
data.head()
```

```python
df = data.groupby('Album_type')[['Likes', 'Views', 'Comments']].mean()
df
```

```python
type(df)
```

```python
df = df.reset_index()
df
```

```python
# Melt - Unpivot a dataframe
df_melted = pd.melt(df, id_vars = 'Album_type', var_name = 'Attribute', value_name = 'Total')
df_melted
```

```python
# Draw the Bar Plot
plt.figure(figsize = (9,4))
sns.barplot(x='Album_type', y='Total', hue='Attribute', data=df_melted)
```

Q5. Find the Top 5 Youtube Channels based on the Views

```python
c_views = data.groupby('Channel')['Views'].sum().sort_values(ascending=False).head()
c_views
```

```python
type(c_views)
```

```python
# sns.set_style("whitegrid")
sns.barplot(x="Views", y="Channel", data=c_views)
plt.title('Top 5 Channels by Views')
plt.xlabel('Views')
plt.ylabel('Channel')
plt.show()
```

Q7. Which Top 7 Tracks have the highest Like-to-View ratio on Youtube ?

```python
data.head()
```

```python
track_lv = data[['Track', 'Likes', 'Views']]
track_lv
```
